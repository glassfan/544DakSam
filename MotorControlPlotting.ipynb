```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
import numpy as np
import pandas as pd
import serial
```

```python
# This set's up the serial connection for the ECE544 Project2 Uartlite connection
ser = serial.Serial()
ser.baudrate = 115200
ser.port = 'COM5'
print(ser)
```

```python
# Create some random data for testing
x = range(0, 100, 1)
y = np.random.rand(100)
```

```python
# setup bokeh to output to the jupyter notebook
output_notebook()
```

```python
# Example Line Plot in Bokeh
p = figure(plot_width=800, plot_height=400, title='Motor Control Plot')
p.line(x, y)
p.xaxis.axis_label = "Sample"
p.yaxis.axis_label = "RPM"
show(p)
```
